```python
import os
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
path = "./test/"
files =[doc for doc in os.listdir(path) if doc.endswith('.txt')]
files
```

```python
contents = []
for file in files:
    with open(f"./test/{file}",'r')as f:
    	temp = f.read()
    	contents.append(temp)
        
contents
```

```python
def vectorize(text):
    model = TfidfVectorizer()
    chk = model.fit_transform(text).toarray()
    return chk
```

```python
vectors = vectorize(contents)
vectors
```

```python
named_vectors = list(zip(files,vectors))
named_vectors
```

```python
def similarity(doc1,doc2):
    return cosine_similarity([doc1,doc2])
```

```python
result = set()
def check():
    global named_vectors
    for name,vect in named_vectors:
        new = named_vectors.copy()
        i = new.index((name,vect))
        del new[i]
        
        for name_b, vect_b in new:
            sim_score  = similarity(vect,vect_b)[0][1]
            sim_score*=100
            file_names = (name,name_b)
            score = (file_names[0],file_names[1],int(sim_score))
            result.add(score)
    return result

check()
```
