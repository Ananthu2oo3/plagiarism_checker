```python
import os
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
path = "./test/"
files =[doc for doc in os.listdir(path) if doc.endswith('.txt')]
files
```

```python
contents = []
for file in files:
    with open(f"./test/{file}",'r')as f:
    	temp = f.read()
    	contents.append(temp)
        
contents
```

```python
def vectorize(text):
    model = TfidfVectorizer()
    chk = model.fit_transform(text).toarray()
    return chk
```

```python
vectors = vectorize(contents)
vectors
```

```python
named_vectors = list(zip(files,vectors))
named_vectors
```

```python
def similarity(doc1,doc2):
    return cosine_similarity([doc1,doc2])
```

```python
result = set()
def check():
    global named_vectors
    for name,vect in named_vectors:
        new = named_vectors.copy()
        i = new.index((name,vect))
        del new[i]
        
        for name_b, vect_b in new:
            sim_score  = similarity(vect,vect_b)[0][1]
            sim_score*=100
            file_names = (name,name_b)
            score = (file_names[0],file_names[1],int(sim_score))
            result.add(score)
    return result
```

```python
data = check()
data
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.DataFrame(data, columns=['File1', 'File2', 'Similarity'])

similarity_matrix = df.pivot(index='File1', columns='File2', values='Similarity')
similarity_matrix = similarity_matrix.fillna(0)
similarity_matrix
```

```python
sns.heatmap(similarity_matrix, annot=True, fmt='g', cmap='coolwarm', linewidths=0.5, cbar_kws={'label': 'Similarity'})
plt.title('Plagiarism Test Result Heatmap')
plt.xlabel('File 2')
plt.ylabel('File 1')
plt.show()
```
